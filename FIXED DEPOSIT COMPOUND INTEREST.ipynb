```python
import plotly.graph_objs as go

# Initialize variables
years = 30
months = years * 12  # Total months
monthly_deposit = 100000
monthly_interest_rate = 0.01
balance = 0
new_balance = []
interest_earned = []

# Iterate over each month
for _ in range(months):
    # Add monthly deposit
    balance += monthly_deposit
    
    # Calculate monthly interest earned
    interest = balance * monthly_interest_rate
    interest_earned.append(interest)
    
    # Add interest to balance
    balance += interest
    new_balance.append(balance)

# Create the balance plot
balance_trace = go.Scatter(
    x=list(range(1, months+1)),
    y=new_balance,
    mode='lines',
    name='Balance',
    hoverinfo='y',
    line=dict(color='blue')
)

balance_layout = go.Layout(
    title='Balance Over Time',
    xaxis=dict(title='Month'),
    yaxis=dict(title='Balance')
)

balance_fig = go.Figure(data=[balance_trace], layout=balance_layout)

# Create the interest plot
interest_trace = go.Scatter(
    x=list(range(1, months+1)),
    y=interest_earned,
    mode='lines',
    name='Interest',
    hoverinfo='y',
    line=dict(color='orange')
)

interest_layout = go.Layout(
    title='Interest Earned Over Time',
    xaxis=dict(title='Month'),
    yaxis=dict(title='Interest')
)

interest_fig = go.Figure(data=[interest_trace], layout=interest_layout)

# Show the plots
balance_fig.show()
interest_fig.show()

# Calculate total deposit, amount earned, and total withdrawals
total_deposit = monthly_deposit * months
total_interest = sum(interest_earned)
total_withdrawal = new_balance[-1]

# Create bar plot
data = [
    go.Bar(x=['Total Deposit', 'Amount Earned', 'Total Withdrawals'],
           y=[total_deposit, total_interest, total_withdrawal],
           marker=dict(color=['blue', 'orange', 'green']))
]

layout = go.Layout(title='Financial Summary',
                   xaxis=dict(title='Category'),
                   yaxis=dict(title='Amount'))

fig = go.Figure(data=data, layout=layout)

# Show the plot
fig.show()
```
